```python
#Importar pacote pandas responsável por carregar e ler o conjunto de dados
import pandas as pd
data = pd.read_csv('wine_dataset.csv')
```

```python
#retornar uma parte do conjunto de dados para visualização 
#head() = função para carregar primeiras linhas do conjunto
#variável sem função = carrega primeiras e últimas linhas do conjunto
data.head()
```

```python
#Substituir o texto por um número, para que o python consiga fazer os futuros cálculos e análises 
data['style'] = data['style'].replace('red', 0)
```

```python
data['style'] = data['style'].replace('white',1)
```

```python
#Separando as variáveis entre preditoras e variável alvo
y = data['style']
```

```python
#Pega todas as colunas, exceto a varivável alvo(style)
#axis=0 (linhas) axis=1 (colunas)
x = data.drop('style', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

#Criação de conjunto de dados para teste e treino
#y = variavel alvo e x = preditoras

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
#Usado para identificar o número de linhas e colunas 
x_treino.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier

#Criação do modelo 
model = ExtraTreesClassifier(n_estimators = 100)
#Os aprndizados serão aplicados a essas variáveis
model.fit(x_treino, y_treino)

#Percentual de aceito 
resultado = model.score(x_teste, y_teste)
print("Resultado:", resultado)
```
